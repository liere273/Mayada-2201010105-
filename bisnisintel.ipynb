## Soal-1
Buatlah table list Top 5 Customers yang memiliki nilai transaksi paling banyak
---





```python
CTop5a=dfc.head(5)
display(CTop5a.set_index("CustomerID"))
```

```python
CTop5b=dfc.loc[:, ["CustomerID", "Amount", "Price"]].groupby("CustomerID").sum().sort_values(by = ['Price'], ascending = False)
CTop5b=CTop5b.rename(columns={"Price":"Gas Consumptions"})
CTop5b.head()
```

## Soal-2
Buatlah list Top 5 Gas stations yang memiliki nilai transaksi paling banyak
---

```python
GSTop5=dfc.loc[:, ["GasStationID", "Amount", "Price"]].groupby("GasStationID").sum().sort_values(by = ['Price'], ascending = False)
GSTop5=GSTop5.rename(columns={"Price":"Gas Sales", "GasStationID":"Gas Station"})
GSTop5.head()
```

## Soal-3
Buatlah list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak
---

```python
PTop5=dfc.loc[:, ["Description", "Amount", "Price"]].groupby("Description").sum().sort_values(by = ['Price'], ascending = False)
PTop5=PTop5.rename(columns={"Price":"Sales"})
PTop5.head()
```
